```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Load the data
heart_data = pd.read_csv('heart.csv')

# Selecting a subset of features for simplicity
df = heart_data[['Age', 'Cholesterol']].dropna()

# Randomly initializing centroids
np.random.seed(200)
k = 3
centroids = {
    i+1: [np.random.randint(df['Age'].min(), df['Age'].max()), np.random.randint(df['Cholesterol'].min(), df['Cholesterol'].max())]
    for i in range(k)
}

# Print randomly generated centroids
print(centroids)

# Plot initial centroids
fig = plt.figure(figsize=(8, 8))
plt.scatter(df['Age'], df['Cholesterol'], color='k')
colmap = {1: 'r', 2: 'g', 3: 'b'}
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(df['Age'].min()-10, df['Age'].max()+10)
plt.ylim(df['Cholesterol'].min()-10, df['Cholesterol'].max()+10)
plt.show()
```

```python
def assignment(df, centroids):
    for i in centroids.keys():
        df['distance_from_{}'.format(i)] = np.sqrt((df['Age'] - centroids[i][0]) ** 2 + (df['Cholesterol'] - centroids[i][1]) ** 2)
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df

df = assignment(df, centroids)

# Plot clusters
fig = plt.figure(figsize=(8, 8))
plt.scatter(df['Age'], df['Cholesterol'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(df['Age'].min()-10, df['Age'].max()+10)
plt.ylim(df['Cholesterol'].min()-10, df['Cholesterol'].max()+10)
plt.show()
```

```python
import copy

old_centroids = copy.deepcopy(centroids)

def update(centroids):
    for i in centroids.keys():
        centroids[i][0] = np.mean(df[df['closest'] == i]['Age'])
        centroids[i][1] = np.mean(df[df['closest'] == i]['Cholesterol'])
    return centroids

centroids = update(centroids)

# Plot updated centroids
fig = plt.figure(figsize=(8, 8))
ax = plt.axes()
plt.scatter(df['Age'], df['Cholesterol'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(df['Age'].min()-10, df['Age'].max()+10)
plt.ylim(df['Cholesterol'].min()-10, df['Cholesterol'].max()+10)
for i in old_centroids.keys():
    old_x = old_centroids[i][0]
    old_y = old_centroids[i][1]
    dx = (centroids[i][0] - old_centroids[i][0]) * 0.75
    dy = (centroids[i][1] - old_centroids[i][1]) * 0.75
    ax.arrow(old_x, old_y, dx, dy, head_width=2, head_length=3, fc=colmap[i], ec=colmap[i])
plt.show()
```

```python
while True:
    closest_centroids = df['closest'].copy(deep=True)
    centroids = update(centroids)
    df = assignment(df, centroids)
    if closest_centroids.equals(df['closest']):
        break

# Plot final clusters
fig = plt.figure(figsize=(8, 8))
plt.scatter(df['Age'], df['Cholesterol'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(df['Age'].min()-10, df['Age'].max()+10)
plt.ylim(df['Cholesterol'].min()-10, df['Cholesterol'].max()+10)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Create KMeans instance
kmeans = KMeans(n_clusters=3)

# Fit the model
kmeans.fit(df[['Age', 'Cholesterol']])

# Get cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Plot clusters
fig = plt.figure(figsize=(8, 8))
plt.scatter(df['Age'], df['Cholesterol'], c=labels, cmap='viridis', alpha=0.5, edgecolor='k')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.xlim(df['Age'].min()-10, df['Age'].max()+10)
plt.ylim(df['Cholesterol'].min()-10, df['Cholesterol'].max()+10)
plt.show()
```
